Define some functions for doing sequential least squares programming.

```python
def minimize(mapping_array, targets, seed, obj_func, **args):

    estimate = scipy.optimize.fmin_slsqp(func = obj_func,x0 = seed,**args)

    fit = mapping_array.dot(estimate[:,numpy.newaxis])[:,0]
     
    return estimate, fit


def nonNegative(mapping_array, targets, seed, obj_func, check_convergence = False, 
                **args):

    def constraints(current_estimate):
        return current_estimate

    estimate, fx, its,imode, smode = scipy.optimize.fmin_slsqp( func = obj_func,
                                                                x0 = seed,
                                                                f_ieqcons = constraints,
                                                                full_output = True,
                                                                **args)
    estimate = numpy.array(estimate)

    if check_convergence:
        assert imode == 0

    fit = mapping_array.dot(estimate[:,numpy.newaxis])[:,0]
   
    return estimate, fit



def leastSquares(mapping_array, targets, seed = None, **args):

    def obj_func(current_estimate):
        current_fit = mapping_array.dot(current_estimate[:,numpy.newaxis])[:,0]    
        return numpy.sum((current_fit - targets) ** 2)

    if seed == None:
        seed = numpy.ones(mapping_array.shape[1], dtype = float)

    return minimize(mapping_array, targets, seed, obj_func, **args)


def nonNegativeLeastMagnitudeDifference(mapping_array, targets, seed = None, check_convergence = False,  **args):

    def obj_func(current_estimate):
        current_fit = mapping_array.dot(current_estimate[:,numpy.newaxis])[:,0]
        return numpy.sum(numpy.absolute((current_fit - targets) / (targets + .01)))

    if seed == None:
        seed = numpy.ones(mapping_array.shape[1], dtype = float)

    return nonNegative(mapping_array, targets, seed, obj_func, check_convergence = check_convergence, **args)


```

Define two classes for simulating a road side detector deployment and estimation, and a utility function for estimation.

```python

def isSubSequence(sub, sup):
    '''Determines if the sequence sub is a possible
    sub-sequence of the sequence sup, where the sub-sequence
    is built by making a binary trial for each item in
    sup and listing the items for which the trial succeeded
    in order. 
    '''

    for x in sub:
        if x not in sup:
            return False

    sup_remain = sup[:]
    for x in sub:
        try:
            i = sup_remain.index(x)
        except ValueError:
            return False
        sup_remain = sup_remain[i+1:]

    return True
        


class Deployment:
    '''Represents a deployment of road-side detectors'''

    def __init__(self, w, detectors, detector_sequences):

        '''
        Parameters:
        
        - `w`: a real number giving the true portion of vehicles that are detectable
        - `detectors`: a mapping keyed by detector name with values that are real
            numbers giving the true probability of a successful detection
        - `detector_sequences`: a mapping with all possible detector sequences as
            keys, and true detector sequence flows as values.
        '''

        self.w = w
        self.detectors = detectors
        self.detector_sequences = detector_sequences

    def simulate(self):
        '''
        Simulates the deployment and returns a dictionary keyed by 
        successful detector sequence, and with values that are the number 
        of times the SDSs occured.
        '''
        SDS_counts = collections.Counter()
        for detector_sequence, count in self.detector_sequences.items():
            
            for trip in range(count):
                if random.random() < self.w:
                    SDS = []
                    for detector in detector_sequence:
                        if random.random() < self.detectors[detector]:
                            SDS.append(detector)

                    if SDS:
                        SDS = tuple(SDS)
                        SDS_counts.update([SDS])

        return SDS_counts

    
    
class Estimator:
    '''Represents something that estimates route flows'''
    def __init__(self, w_guesser = None, P_guesser = None):
        '''
        Parameters:

        - `w_guesser`: a function that takes the true value of w and 
            returns a guessed value of w
        - `w_guesser`: a function that takes the true value of P and 
            returns a guessed value of P
        '''
        self.w_guesser = w_guesser
        self.P_guesser = P_guesser


    def estimate(self, w, detectors, detector_sequences, SDS_counts):
        '''
        Estimate the route flows.

        Parameters:

        - `w`: a real number giving the true portion of vehicles that are detectable
        - `detectors`: a mapping keyed by detector name with values that are real
            numbers giving the true probability of a successful detection
        - `detector_sequences`: a sequence all possible detector sequences, each
            of which is a tuple giving a sequence of detector names.
        '''
        
        #get w and P with uncertainty/error
        estimated_w = self.w_guesser(w)
        estimated_P = self.P_guesser(detectors)

        del w
        del detectors
        
        
        
        #get DSs
        DSs = tuple(sorted(detector_sequences))

        #get the SDSs
        SDSs = set()
        for ds in DSs:
            for r in range(1, len(ds)+1):
                for sds in itertools.combinations(ds, r):
                    SDSs.add(sds)
                
        SDSs = tuple(sorted(SDSs))
    
        #get the design matrix
        A = numpy.zeros(shape = (len(SDSs), len(DSs)))
    
        
    
        for row, sds in enumerate(SDSs):
            for col, ds in enumerate(DSs):
                if not isSubSequence(sds, ds):
                    p = 0.
                else:
                    p = 1.
                for d in ds:
                    if d in sds:
                        p *= estimated_P[d]
                    else:
                        p *= (1.-estimated_P[d])
                A[row,col] = p * estimated_w
        
        #get the target
        B = numpy.zeros(shape = (len(SDSs),))
        for row, sds in enumerate(SDSs):
            B[row] = SDS_counts[sds]

        #make the initial guess at the solution
        initial_guess = leastSquares(A,
                                              B,
                                              [100.] * A.shape[1],
                                              iter = 100, iprint = 0)[0]

        #find the final solution
        solution, fit = nonNegativeLeastMagnitudeDifference(A,
                                                                     B,
                                                                     initial_guess,
                                                                     check_convergence = True,
                                                                     iter = 10000,
                                                                     iprint = 0)

        
        
        return dict(zip(DSs, solution)), estimated_w, estimated_P
    
    
    def bootstrap(self, w, detectors, detector_sequences, num_sims = 100):
        

        
        DSs = tuple(sorted(detector_sequences))

        sims = numpy.zeros(shape = (len(DSs),num_sims))
        
        for k,v in detector_sequences.items():
            detector_sequences[k] = int(round(v))
        
        deployment = Deployment(w, detectors, detector_sequences)
        
        
        for i in range(num_sims):
            
            SDS_counts = deployment.simulate()
            solution = self.estimate(w, detectors, detector_sequences.keys(), SDS_counts)[0]
            solution = [solution[ds] for ds in DSs]
            sims[:,i] = solution
            
        
        qs = scipy.stats.mstats.mquantiles(sims, prob=[0.025, .975], axis = 1)
        assert qs.shape == (len(DSs), 2)
        
        return dict(zip(DSs, qs))
```
